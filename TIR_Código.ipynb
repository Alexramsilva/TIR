```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definición de una función en Python
def f(r):
    return (-1000/(1+r)**0+300/(1+r)**1+400/(1+r)**2+550/(1+r)**3)
```

```python
# Calculamos el dominio de la función f(x)
r= np.linspace(0, 100, 101)/100
```

```python
f1=f(r)
```

```python
# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(r,f1, label=f'Cash Flow', color='#e7d84b')

plt.title('Gráfico tasa de interés')
plt.xlabel('r')
plt.ylabel('f(r)')
plt.grid(True)
plt.legend()
plt.axhline(0, color='red',linewidth=0.9)
plt.axvline(0, color='yellow',linewidth=0.9)
plt.show()
```

```python
# Función que representa el valor presente neto (VPN) de los flujos de efectivo
def vp(r, FE):
    vp = 0
    for t, fe in enumerate(FE):
        vp += fe / (1 + r) ** t
    return vp
```

```python
# Función para calcular la derivada del VPN con respecto a la tasa
def derivada(r, FE):
    derivada = 0
    for t, fe in enumerate(FE):
        derivada += -t * fe / (1 + r) ** (t + 1)
    return derivada
```

```python
# Flujos de efectivo (reemplaza estos valores con los de tu proyecto)
FE = [-1000, 300, 400, 550]
```

```python
# Supongamos una tasa inicial cercana a la TIR real
r_0 = 0.5
max_iteracciones = 100
tolerancia = 1e-6
```

```python
# Método de Newton-Raphson para encontrar la TIR
for i in range(max_iteracciones):
    vp_valor = vp(r_0, FE)
    vp_derivada = derivada(r_0, FE)

    r_0 = r_0 - vp_valor / vp_derivada

    if abs(vp_valor) < tolerancia:
        break
```

```python
# Verificar si se encontró una solución
if i < max_iterations:
    print(f'Tasa Interna de Retorno (TIR): {r_0:.2%}')
else:
    print('No se encontró una solución después de', max_iteracciones, 'iteracciones.')

```
